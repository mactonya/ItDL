## 1. ndarray concept and tensor operation

```python
#!pip install numpy
import numpy as np
""" 1D ndArray """
a_1d = [1, 2, 3, 4]
b_1d = np.array(a_1d)
print(type(b_1d))
print(b_1d.shape)
print(b_1d)
```

```python
""" 2D ndArray """
a_2d = [[1, 2, 3, 4],
        [5, 6, 7, 8]]
b_2d = np.array(a_2d)
print(type(b_2d))
print(b_2d.shape)
print(b_2d)
```

```python
""" 3D ndArray """
a_3d = [[[1, 2],
        [3, 4]],
        
        [[5, 6],
        [7, 8]]]
b_3d = np.array(a_3d)
print(type(b_3d))
print(b_3d.shape)
print(b_3d)
```

```python
""" np.multiple """
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[1, 2],
              [3, 4]])
c = np.multiply(a, b)
print("a")
print(a)
print("b")
print(b)
print("c = np.multiply(a, b)")
print(c)
```

```python
""" np.dot """
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[1, 2],
              [3, 4]])
c = np.dot(a, b) # c = a @ b
print("a")
print(a)
print("b")
print(b)
print("c = np.dot(a, b) or c = a @ b")
print(c)
```

```python
""" np.add """
a = np.array([[1, 2],
              [3, 4]])
b = np.array([[1, 2],
              [3, 4]])
c = np.add(a, b) # c = a + b
print("a")
print(a)
print("b")
print(b)
print("c = np.add(a, b) or c = a + b")
print(c)
```

```python
""" np.transpose """
a = np.array([[1, 2],
              [3, 4]])
b = np.transpose(a)
print("a")
print(a)
print("b = np.transpose(a) or b = a.T" )
print(b)
print()
```

```python
""" np.zeros """
shape = (3,2,1)
a = np.zeros(shape)
print(a.shape)
print(a)
```

```python
""" np.reshape """
a = np.array([1, 2, 3, 4, 5, 6])
b = np.reshape(a, (2, -1))
print(a.shape)
print(b.shape)
```

```python
""" np.stack """
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.stack((a,b), axis = 0)
print(c)
d = np.stack((a,b), axis = 1)
print(d)
```

## 2. plot the picture based on "matplotlib library"

```python
import matplotlib.pyplot as plt
import numpy as np

""" basic """
x = np.linspace(-1, 1, 50)
y = 2*x + 1
plt.figure(num=0)
plt.plot(x, y)
plt.show()
```

```python
""" advance """
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2

plt.figure(num=1, figsize=(8, 5),)
plt.title('I am title', size=18)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.show()
```

```python
""" set label and limit"""
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure(num=2, figsize=(8, 5),)
plt.plot(x, y2)
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--')
plt.xlim((-1, 2))
plt.ylim((-2, 3))
plt.xlabel('I am x')
plt.ylabel('I am y')
plt.show()
```

```python
""" set legend """
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2, label='linear line')
plt.plot(x, y1, color='red', linewidth=1.0, linestyle='--', label='square line')
plt.legend(loc='upper right')
plt.show()
```

```python
""" scatter picture"""
n = 1024    # data size
X = np.random.normal(0, 1, n) 
Y = np.random.normal(0, 1, n) 
T = np.arctan2(Y,X) # for color value

plt.figure(num=4, figsize=(8, 5),)
plt.scatter(X, Y, s=75, c=T, alpha=.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.show()
```

```python
""" bar picture"""
n = 100
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0, 1.0, n)

plt.figure(num=5, figsize=(8, 5),)
plt.bar(X, Y1)
plt.xlim(-.5, n)
plt.ylim(0, 1.25)

plt.show()
```

```python
""" animate """
%matplotlib inline
from matplotlib import animation, rc
from IPython.display import HTML

# First set up the figure, the axis, and the plot element we want to animate
fig, ax = plt.subplots()

ax.set_xlim(( 0, 2))
ax.set_ylim((-2, 2))

line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)

# animation function. This is called sequentially
def animate(i):
    x = np.linspace(0, 2, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return (line,)

# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=20, blit=True)

plt.close()
HTML(anim.to_jshtml())
```

## 3. basic file I/O(read image/csv file)

```python
#!pip install Pillow
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

""" From image to array """
image = Image.open("example_pic/example.jpg")
image2array = np.array(image)
print(image2array.shape)
```

```python
""" From array to image """
array2image = Image.fromarray(image2array, mode = 'RGB')
array2image.save("example_pic/array2image.jpg")
plt.figure(num=6, figsize=(8, 5),)
plt.imshow(array2image)
```

```python
""" reshape """
image = Image.open("example_pic/example.jpg")
reshape_image = image.resize((300,300), Image.BILINEAR)
reshape_image.save("example_pic/reshape_image.jpg")
plt.figure(num=7, figsize=(8, 5),)
plt.imshow(reshape_image)
```

```python
""" rotate """
image = Image.open("example_pic/example.jpg")
rotate_image = image.rotate(30, Image.BILINEAR)
rotate_image.save("example_pic/rotate_image.jpg")
plt.figure(num=8, figsize=(8, 5),)
plt.imshow(rotate_image)
```
